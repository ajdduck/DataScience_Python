```python
df = pd.read_csv('data/titanic.csv', index_col = 0)
df
```

```python
age = df['Age']
age
```

```python
sns.distplot(age, bins=10)
```

```python
df[['Age', 'Fare']]
```

```python
df['Fare'].max()
```

```python
max_fare = df['Fare'] == 512.3292
```

```python
df.loc[max_fare, 'Age']
```

```python
df.head()
```

```python
df['Survived'].value_counts()
```

```python
df['Pclass'].value_counts()
```

```python
for i in range(1,4):
    survived_counts = df.loc[df['Pclass'] == i, 'Survived'].value_counts()
    
    
    print(survived_counts[1]/(survived_counts[0]+survived_counts[0]))
    
```

```python
sns.heatmap(df[['Survived', 'Age']].corr(), annot = True) 
# -0.077이 유의미함을 보여주기에는 어려운 수치라고 할 수 있음 (틀린 풀이)
```

```python
survived_df = df[['Survived', 'Age', 'Sex']]
```

```python
survived_df
```

```python
for i in range(1, 892):
    if survived_df['Sex'][i] == 'male':
        survived_df['Sex'][i] = 0
    else:
        survived_df['Sex'][i] = 1

survived_df
        
```

```python
survived_df = survived_df.apply(pd.to_numeric, errors = 'coerce')
```

```python
survived_df.corr()
```

##### 모범 답안

```python
titanic_df = pd.read_csv('data/titanic.csv', index_col = 0)
titanic_df.info()
```

```python
# 나이분포

titanic_df.plot(kind='hist', y='Age', bins = 10)
```

```python
# 나이별 요금 분포
# 산점도 두 변수 간의 상관관계를 파악하기 위함.

titanic_df.plot(kind='scatter', x='Age', y='Fare')
```

```python
# 생존 여부 파악

titanic_df['Survived'].value_counts()
```

```python
# 객실 등급별 탑승인원

titanic_df['Pclass'].value_counts()
```

```python
# 객실 등급별 생존율 분포

titanic_df.plot(kind='scatter', x='Pclass', y='Survived')
```

```python
sns.kdeplot(titanic_df['Pclass'], titanic_df['Survived'])
```

```python
# 생존 여부에 따른 나이 분포

# 카테고리별 분류

sns.stripplot(data=titanic_df, x='Survived', y='Age')
```

```python
sns.violinplot(data=titanic_df, x='Survived', y='Age')
```

```python
# 생존 여부에 따른 나이 및 성별 분포
# hue > 자료형 object 반영

sns.stripplot(data=titanic_df, x='Survived', y='Age', hue='Sex')
```
