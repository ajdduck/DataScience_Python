#### 연습 문제 2

```python
df = pd.read_csv('data/enrolment_1.csv')

# 코드를 작성하세요.
df['status'] = 'allowed'

condition1 = (df['year'] == 1) &(df['course name'] == 'infromation technology')
condition2 = (df['year'] == 4) & (df['course name'] == 'commerce')

xcourse_names = df['course name'].value_counts() # dtype이 int64임 
condition3 = xcourse_names < 5

xcourse_names_list = list(xcourse_names[condition3].index) 
# 즉, 출력된 데이터는 정수이고 이 데이터의 인덱싱 위치가 곧 course name이 되는 것.

for xcourse in xcourse_names_list:
    df.loc[df['course name'] == xcourse, 'status'] = 'not allowed'

df.loc[condition1, 'status'] = 'not allowed'
df.loc[condition2, 'status'] = 'not allowed'


# 정답 출력
df
```

#### 연습 문제 3

```python
df = pd.read_csv('data/enrolment_2.csv')

# 코드를 작성하세요.


# 80명 이상의 학생이 수강하는 과목 'Auditorium'

allowed = df['status'] == 'allowed'

course_counts = df.loc[df['status'] == 'allowed', 'course name'].value_counts()
eighty_course_list = list(course_counts[course_counts > 80].index)

for eighty_course in eighty_course_list:
    df.loc[(df['course name'] == eighty_course) & allowed, 'room assignment'] = 'Auditorium'

# 40    ,"Large" room"
forty_course_list = list(course_counts[(course_counts >= 40) & (course_counts < 80)].index)

for forty_course in forty_course_list:
    df.loc[(df['course name'] == forty_course) & allowed, 'room assignment'] = 'Large room'

# 15 ~ 40    , "Medium room"
fifty_course_list = list(course_counts[(course_counts >= 15) & (course_counts < 40)].index)

for fifty_course in fifty_course_list:
    df.loc[(df['course name'] == fifty_course) & allowed, 'room assignment'] = 'Medium room'

# 5 ~ 15    , "Small room"
five_course_list = list(course_counts[(course_counts >=5) & (course_counts < 15)].index)

for five_course in five_course_list:
    df.loc[(df['course name'] == five_course) & allowed, 'room assignment'] = 'Small room'

# 'not allowed' > 'not assigned'
not_allowed = df['status'] == 'not allowed'
df.loc[not_allowed, 'room assignment'] = 'not assigned'

# 정답 출력
df
```

#### 연습 문제 4

```python
df = pd.read_csv('data/enrolment_3.csv')

# 코드를 작성하세요.

allowed = df['status'] == 'allowed'
room_assignment1 = allowed & (df['room assignment'] == 'Auditorium')
auditorium_list=list(df.loc[room_assignment1, 'course name'].unique())
auditorium_list.sort()

room_assignment2 = allowed & (df['room assignment'] == 'Large room')
large_list = list(df.loc[room_assignment2, 'course name'].unique()) # Series 자료형
large_list.sort()

room_assignment3 = allowed & (df['room assignment'] == 'Medium room')
medium_list = list(df.loc[room_assignment3, 'course name'].unique()) # Series 자료형
medium_list.sort()

room_assignment4 = allowed & (df['room assignment'] == 'Small room')
small_list = list(df.loc[room_assignment4, 'course name'].unique()) # Series 자료형
small_list.sort()

for course in auditorium_list:
    i = auditorium_list.index(course) + 1
    df.loc[(df['course name'] == course) & allowed, 'room assignment'] = 'Auditorium-%d' % i
    
for course in large_list:
    i = large_list.index(course) + 1
    df.loc[(df['course name'] == course) & allowed, 'room assignment'] = 'Large-%d' % i    

for course in medium_list:
    i = medium_list.index(course) + 1
    df.loc[(df['course name'] == course) & allowed, 'room assignment'] = 'Medium-%d' % i

for course in small_list:
    i = small_list.index(course) + 1
    df.loc[(df['course name'] == course) & allowed, 'room assignment'] = 'Small-%d' % i

df.rename(columns={'room assignment': 'room number'}, inplace = True)

df
```
