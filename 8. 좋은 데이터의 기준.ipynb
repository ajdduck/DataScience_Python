#### 데이터 클리닝: 정확성

이상점, 다른값들과 지나치게 동떨어져있는 값

판단하는 기준은?

1.5*IQR(interquartile Range)에서 벗어난 값 Box Plot(25% ~ 75%)

이상점 처리도 결측치 처리와 판단을 비슷하게.

```python
%matplotlib inline
import pandas as pd
```

```python
df = pd.read_csv('data/beer.csv', index_col = 0)
df.head()
```

```python
df.plot(kind='box', y = 'abv')
```

```python
df['abv'].describe()
```

```python
df['abv'].quantile(0.25)
```

```python
df['abv'].quantile(0.75)
```

```python
q1 = df['abv'].quantile(0.25)
q2 = df['abv'].quantile(0.75)

iqr = q2 - q1
```

```python
condition = (df['abv'] < q1 - 1.5*iqr)|(df['abv']> q2 + 1.5*iqr)
```

```python
df[condition]
```

```python
df.loc[2250, 'abv'] = 0.055
df.loc[2250]
```

```python
condition = (df['abv'] < q1 - 1.5*iqr)|(df['abv']> q2 + 1.5*iqr)
df[condition]
```

```python
df[condition].index
```

```python
df.drop(df[condition].index, inplace = True)
```

```python
df.head()
```

```python
condition = (df['abv'] < q1 - 1.5*iqr)|(df['abv']> q2 + 1.5*iqr)
df[condition]
```

```python
df.plot(kind='box', y = 'abv')
```

#### 연습 문제

```python
df = pd.read_csv('data/movie_metadata.csv')

# 코드를 작성하세요.

df.plot(kind = 'scatter', x = 'budget', y= 'imdb_score')
```

```python
q1 = df['budget'].quantile(0.25)
q2 = df['budget'].quantile(0.75)

iqr = q2 - q1

# 이상점 확인, 어느 수준을 이상점으로 할 것인가에 대한 주체적인 판단도 필요

condition = df['budget'] > q2 + 5*iqr
df.drop(df[condition].index, inplace = True)

df.plot(kind = 'scatter', x = 'budget', y='imdb_score')
```
